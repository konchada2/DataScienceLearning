### Difference between iloc and loc

__iloc__ doesnt work when we actually call the column name. Its only accesible by its index. When we have large data and we dont the column number then __loc__ can come in handy. 

```python
df.loc[df['FirstName'] == 'Fernanda', 'City'] # giving city of the person called fernanda (Filtering)
```

```python
df.set_index('LastName', inplace=True) # setting a particular column as index. we can view it as a primary key.
```

```python
df.head()
```

```python
#Select the row where 'Andrade' is present

df.loc['Andrade']
```

```python
df.loc['Andrade', 'Address']
```

```python
df.loc[['Andrade', 'Veness'], 'City']
```

```python
df.loc[df['County'] == 'Kent', :] # filtering all the records where county is kent
```

```python
df.loc[df['FirstName'] == 'France', ['City', 'Web', 'County'] ]
```

```python
df.loc[df['Email'].str.endswith('gmail.com')]
```

```python
#Multiple conditions in rows
df.loc[df['FirstName'].isin(['France', 'Tyisha', 'Eric'])]
```

```python
#Range of columns
df.loc[df['FirstName'].isin(['France', 'Tyisha', 'Eric']), 'Address': 'Email'] # from address to email 
```

```python
#Subsetting based on condition
df.loc[df['Company'].apply(lambda x : len(x.split(' ')) == 4)] # where name of the company has 4 words
```
